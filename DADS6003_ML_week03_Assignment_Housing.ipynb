```python
import pandas as pd

file_path = r'/content/drive/MyDrive/Colab Notebooks/DADS6003_ML_week03_Dataset_Housing.csv'
housing = pd.read_csv(file_path)
housing.head()
```

### **Data review**

```python
print(f'Information:\n{housing.info()}')
print('-----------------------------------------------------------')
print(f'Description:\n{housing.describe()}')
print('-----------------------------------------------------------')
print(f'Check null:\n{housing[housing.isnull()].count()}')
```

### **Data preparation**

```python
housing.iloc[:,5:12] = housing.iloc[:,5:12].replace({'yes':1,'no':0})
housing['furnishingstatus'] = housing['furnishingstatus'].replace({'furnished':3,'semi-furnished':2,'unfurnished':1})
housing.head()

#plt.figure(figsize=(14,8))
#sns.set_theme(style="white")
#corr = housing.corr()
#heatmap = sns.heatmap(corr, annot=True, cmap="Blues", fmt='.1g')
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing, test_size = 0.2, random_state = 42)
print('Training set')
display(train_set.head())
print('-------------------------------------------------------------------------\nTest set')
display(test_set.head())
print('-------------------------------------------------------------------------\nCorrelation')
corr_matrix = housing.corr()
print(corr_matrix['price'].sort_values(ascending = False))
```

### **Linear model**

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler, StandardScaler

x = train_set.iloc[:, 1:-1]
y = train_set.iloc[:, 0]

minmax_scaler = MinMaxScaler()
std_scaler = StandardScaler()

#Feature scaling
x_minmax_scaled = pd.DataFrame(minmax_scaler.fit_transform(x))
x_std_scaled = pd.DataFrame(std_scaler.fit_transform(x))

#Unscaled Train model
Lin_reg = LinearRegression()

lin_reg = Lin_reg.fit(x, y)
print(f'Intercept:\t{lin_reg.intercept_}\n')
print(f'Coefficient:\t{lin_reg.coef_}\n')

#Unscaled Predict
x_test = test_set.iloc[:, 1:-1]
y_test = test_set.iloc[:, 0]
#display(x_test)

x_unscaled_results = lin_reg.predict(x_test)
#print(x_unscaled_results)
mse_lin_unscaled = sklearn.metrics.mean_squared_error(y_test, x_unscaled_results)
mae_lin_unscaled = sklearn.metrics.mean_absolute_error(y_test, x_unscaled_results)
r2_lin_unscaled = sklearn.metrics.r2_score(y_test, x_unscaled_results)
print(f'Mean square error linear unscaled:\t{mse_lin_unscaled}\n\
Mean absolute error linear unscaled:\t{mae_lin_unscaled}\n\
R2 linear unscaled:\t\t\t{r2_lin_unscaled}\n----------------------------------------\n')

```

### **ทำ Feature scaling กับ Linear model**

```python
#MinMaxscaled Train model
mm_lin_reg = Lin_reg.fit(x_minmax_scaled, y)
print(f'Intercept of minmax scaled:\t{mm_lin_reg.intercept_}\n')
print(f'Coefficient of minmax scaled:\t{mm_lin_reg.coef_}\n')

#MinMaxscaled Predict
x_test_mm = minmax_scaler.transform(x_test)
#display(pd.DataFrame(x_test_mm))

x_mmscaled_results = mm_lin_reg.predict(x_test_mm)
#print(x_mmscaled_results)
mae_lin_mmscaled = sklearn.metrics.mean_absolute_error(y_test, x_mmscaled_results)
r2_lin_mmscaled = sklearn.metrics.r2_score(y_test, x_mmscaled_results)
print(f'Mean absolute error linear Min-Max scaled:\t{mae_lin_mmscaled}\n\
R2 linear Min-Max scaled:\t\t\t{r2_lin_mmscaled}\n----------------------------------------\n')

#Stdcaled Train model
std_lin_reg = Lin_reg.fit(x_std_scaled, y)
print(f'Intercept of std scaled:\t{std_lin_reg.intercept_}\n')
print(f'Coefficient of std scaled:\t{std_lin_reg.coef_}\n')

#MinMaxscaled Predict
x_test_std = std_scaler.transform(x_test)
#display(pd.DataFrame(x_test_std))

x_stdscaled_results = std_lin_reg.predict(x_test_std)
#print(x_stdscaled_results)
mae_lin_stdscaled = sklearn.metrics.mean_absolute_error(y_test, x_stdscaled_results)
r2_lin_stdscaled = sklearn.metrics.r2_score(y_test, x_stdscaled_results)
print(f'Mean absolute error linear STD scaled:\t{mae_lin_mmscaled}\n\
R2 linear STD scaled:\t\t\t{r2_lin_stdscaled}\n----------------------------------------\n')
```

### **ทดลองใช้ Pipeline ทำ Feature scaling กับ Linear model**



```python
from sklearn.pipeline import Pipeline

pipe_lin_minmax_scaled = Pipeline([
    ('scaler', MinMaxScaler()),
    ('lr', LinearRegression())
    ])

display(pipe_lin_minmax_scaled.fit(x, y))
print(f'Intercept of lin minmax:\t{pipe_lin_minmax_scaled.named_steps["lr"].intercept_}\n')
print(f'Coefficient of lin minmax:\t{pipe_lin_minmax_scaled.named_steps["lr"].coef_}\n')
lin_results = pipe_lin_minmax_scaled.predict(x_test)
mae_poly = sklearn.metrics.mean_absolute_error(y_test, lin_results)
r2_poly = sklearn.metrics.r2_score(y_test, lin_results)
print(f'Mean absolute error linear minmax:\t{mae_poly}\n\
R2 linear minmax:\t\t\t{r2_poly}\n----------------------------------------\n')

pipe_lin_std_scaled = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LinearRegression())
    ])

display(pipe_lin_std_scaled.fit(x, y))
print(f'Intercept of lin std:\t{pipe_lin_std_scaled.named_steps["lr"].intercept_}\n')
print(f'Coefficient of lin std:\t{pipe_lin_std_scaled.named_steps["lr"].coef_}\n')
lin_results = pipe_lin_std_scaled.predict(x_test)
mae_poly = sklearn.metrics.mean_absolute_error(y_test, lin_results)
r2_poly = sklearn.metrics.r2_score(y_test, lin_results)
print(f'Mean absolute error linear std:\t{mae_poly}\n\
R2 linear std:\t\t\t{r2_poly}\n----------------------------------------\n')
```

### **Polynomial model**

```python
from sklearn.preprocessing import PolynomialFeatures

pipe_poly_unscaled = Pipeline([
    ('scaler', PolynomialFeatures(degree=2)),
    ('lr', LinearRegression())
    ])

display(pipe_poly_unscaled.fit(x, y))
print(f'Intercept of poly:\t{pipe_poly_unscaled.named_steps["lr"].intercept_}\n')
print(f'Coefficient of poly:\t{pipe_poly_unscaled.named_steps["lr"].coef_}\n')
poly_results = pipe_poly_unscaled.predict(x_test)
mae_poly = sklearn.metrics.mean_absolute_error(y_test, poly_results)
r2_poly = sklearn.metrics.r2_score(y_test, poly_results)
print(f'Mean absolute error poly:\t{mae_poly}\n\
R2 poly:\t\t\t{r2_poly}\n----------------------------------------\n')
```

### **ทดลองทำ Feature scaling กับ Polynomial model**

```python
pipe_poly_minmaxscaled = Pipeline([
    ('pre-scaler', MinMaxScaler()),
    ('scaler', PolynomialFeatures(degree=2)),
    ('lr', LinearRegression())
    ])

display(pipe_poly_minmaxscaled.fit(x, y))
print(f'Intercept of poly minmax:\t{pipe_poly_minmaxscaled.named_steps["lr"].intercept_}\n')
print(f'Coefficient of poly minmax:\t{pipe_poly_minmaxscaled.named_steps["lr"].coef_}\n')
poly_results = pipe_poly_minmaxscaled.predict(x_test)
mae_poly = sklearn.metrics.mean_absolute_error(y_test, poly_results)
r2_poly = sklearn.metrics.r2_score(y_test, poly_results)
print(f'Mean absolute error poly minmax:\t{mae_poly}\n\
R2 poly minmax:\t\t\t{r2_poly}\n----------------------------------------\n')

pipe_poly_stdscaled = Pipeline([
    ('pre-scaler', StandardScaler()),
    ('scaler', PolynomialFeatures(degree=2)),
    ('lr', LinearRegression())
    ])

display(pipe_poly_stdscaled.fit(x, y))
print(f'Intercept of poly std:\t{pipe_poly_stdscaled.named_steps["lr"].intercept_}\n')
print(f'Coefficient of poly std:\t{pipe_poly_stdscaled.named_steps["lr"].coef_}\n')
poly_results = pipe_poly_stdscaled.predict(x_test)
mae_poly = sklearn.metrics.mean_absolute_error(y_test, poly_results)
r2_poly = sklearn.metrics.r2_score(y_test, poly_results)
print(f'Mean absolute error poly std:\t{mae_poly}\n\
R2 poly std:\t\t\t{r2_poly}\n----------------------------------------\n')
```
